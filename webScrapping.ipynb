```python
import numpy as np
import pandas as pd
import requests as rq
!pip install bs4
from bs4 import BeautifulSoup as bs
```

```python
all_authors = []
all_tags = []
all_quotes = []

for i in range(1, 11):
    url = f'https://quotes.toscrape.com/page/{i}/'
    res = rq.get(url)
    soup = bs(res.text, 'html.parser')

    # Extract authors
    current_authors = soup.find_all('small')
    all_authors.extend([a.text for a in current_authors])

    # Extract quotes
    current_quotes = soup.find_all('span', class_='text')
    # Filter quotes that are long enough, as per previous logic
    all_quotes.extend([q.text for q in current_quotes if len(q.text) > 10])

    # Extract tags
    divs = soup.find_all('div', class_='quote')
    for div in divs:
        current_tags_in_div = div.find_all('a', class_='tag')
        all_tags.append([t.text for t in current_tags_in_div])

# Now, create the DataFrame using the collected data
df_all_pages = pd.DataFrame(data=[all_authors, all_tags, all_quotes], dtype=str).T
df_all_pages.columns = ['authors', 'tags', 'quotes']

print(f"Shape of the DataFrame after collecting from all pages: {df_all_pages.shape}")
display(df_all_pages.head())
```
